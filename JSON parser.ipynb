```python
from collections import namedtuple
import json
geo_info_list = []

class Geometry:
    def __init__(self, coordinates):
        self.type = "Point"
        self.coordinates = coordinates
    def reprJSON(self):
        return dict(type = self.type, coordinates = self.coordinates) 

class Properties:
    def __init__(self, name):
        self.name = name
    def reprJSON(self):
        return dict(name = self.name) 

class Doc:
    def __init__(self, geom, prop):
        self.type = "Feature"
        self.geometry = geom
        self.properties = prop
    def reprJSON(self):
        return dict(type = self.type, geometry = self.geometry, properties = self.properties) 

class ComplexEncoder(json.JSONEncoder):
    def default(self, obj):
        if hasattr(obj,'reprJSON'):
            return obj.reprJSON()
        else:
            return json.JSONEncoder.default(self, obj)

with open('Noise.json') as json_file:
    data = json.load(json_file)
    for place in data['places']:
        geom = Geometry([place['longitude'],place['latitude']])
        prop = Properties(place['address'])
        doc = Doc(geom, prop)
        geo_info_list.append(doc)

        

#print("Full JSON")
with open('Noise.js', 'w') as outfile:
    for g in geo_info_list:
        json_str = json.dumps(g.reprJSON(), cls=ComplexEncoder);
        outfile.write(json_str + ',\n');
    #print("Partial JSON")
    #json_string = json.dumps([geo.__dict__ for geo in geo_info_list])
    
```
